# PARTE 3: Limpieza de Datos del DataFrame combinado

- Objetivo: Arreglar Datos problemáticos + Tratar Nulos 

```python
# PASO 1: Solucionar los números negativos de la columna "Salary" vistos en el "min" de la función "descirbe"

df_combinado.columns
```

```python
df_combinado["Salary"].sample(3) # A simple vista no se ve claro que existan números negativos
```

```python
salarios_negativos = df_combinado[df_combinado["Salary"] < 0] # Filtramos para comporbar mejor la existencia de los datos negativos 

print(f"Número de registros con salario negativo: {len(salarios_negativos)}")

salarios_negativos.sample(3)
```

```python
df_combinado["Salary"] = df_combinado["Salary"].abs() # Transfromamos los número snegativos en números positivos
```

```python
salarios_negativos_final = df_combinado[df_combinado["Salary"] < 0] # Comprobamos los números negativos ahora = 0

print(f"Número de registros con salario negativo: {len(salarios_negativos_final)}")
```

```python
# PASO 2: Tratar los Nulos de las columnas "Cancellation Year", "Cancellation Month" y "Salary"

round(df_combinado.isna().sum().sort_values(ascending=False)/df_flight.shape[0]*100,2)
```
